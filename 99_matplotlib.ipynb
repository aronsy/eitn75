To use matplotlib and show plots in a notebook there are different ways; two (2) are shown here.

1. __%matplotlib notebook__ will give you an interactive plot, but you MUST remeber to shut it down before plotting the next.
2. __%matplotlib inline__ will just generate an inline plot, no interactive tools.

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np

def plot_sine():
    x = np.linspace(0, 2*np.pi, 10)
    y = np.sin(x)
    plt.plot(x, y)
```

```python
plot_sine()
```

```python
%matplotlib inline

plot_sine()
```
