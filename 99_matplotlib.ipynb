## Using matplotlib in notebooks
To use matplotlib and show plots in a notebook there are different ways; two (2) are shown here.

1. __%matplotlib notebook__ will give you an interactive plot, but you MUST remeber to shut it down before plotting the next.
2. __%matplotlib inline__ will just generate an inline plot, no interactive tools. You should also set the size: plt.rcParams['figure.figsize'] = [9.5, 6]

````python
import matplotlib.pyplot as plt
import numpy as np

def plot_sine():
    x = np.linspace(0, 2*np.pi, 100)
    y1 = np.sin(x)
    y2 = np.sin(x + np.pi/2)
    plt.plot(x, y1, label='sin(x)')
    plt.plot(x, y2, label='cos(x)')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.grid(True)
    plt.legend()
````

````python
%matplotlib notebook
plot_sine()
````

````python
%matplotlib inline
plt.rcParams['figure.figsize'] = [12, 6]

plot_sine()
````
